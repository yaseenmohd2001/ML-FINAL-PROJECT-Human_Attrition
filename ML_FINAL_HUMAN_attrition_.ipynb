```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Datasets/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df
```

```python
cols_to_drop = [
    # Useless or constant columns
    'EmployeeCount',
    'Over18',
    'StandardHours',
    'EmployeeNumber',
    'PerformanceRating',
    'DailyRate',
    'HourlyRate',
    'MonthlyRate',

    # Redundant or confusing columns
    'NumCompaniesWorked',
    'PercentSalaryHike',
    'StockOptionLevel',
    'TrainingTimesLastYear',
    'YearsSinceLastPromotion',
    'YearsWithCurrManager',

    # Highly subjective satisfaction metrics
    'EnvironmentSatisfaction',
    'RelationshipSatisfaction',
    'JobInvolvement',
    'WorkLifeBalance'
]
```

```python
df.drop(columns=cols_to_drop,inplace=True)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
import pickle

# Initialize separate encoders for each categorical column
le_attrition = LabelEncoder()
le_businesstravel = LabelEncoder()
le_department = LabelEncoder()
le_educationfield = LabelEncoder()
le_gender = LabelEncoder()
le_jobrole = LabelEncoder()
le_maritalstatus = LabelEncoder()
le_overtime = LabelEncoder()

# Apply encoding to categorical columns
df['Attrition'] = le_attrition.fit_transform(df['Attrition'])
df['BusinessTravel'] = le_businesstravel.fit_transform(df['BusinessTravel'])
df['Department'] = le_department.fit_transform(df['Department'])
df['EducationField'] = le_educationfield.fit_transform(df['EducationField'])
df['Gender'] = le_gender.fit_transform(df['Gender'])
df['JobRole'] = le_jobrole.fit_transform(df['JobRole'])
df['MaritalStatus'] = le_maritalstatus.fit_transform(df['MaritalStatus'])
df['OverTime'] = le_overtime.fit_transform(df['OverTime'])
```

```python
df
```

```python
X=df.drop(columns="Attrition")
X
```

```python
y=df["Attrition"]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report
knn = KNeighborsClassifier()
svc = SVC()
nb = GaussianNB()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
ada = AdaBoostClassifier()
gboost = GradientBoostingClassifier()
xgb = XGBClassifier()
models = [knn,svc,nb,dt,rf,ada,gboost,xgb]
for model in models:
  print(model)
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  print(classification_report(y_test,y_pred))
  print("*"*100)
```
